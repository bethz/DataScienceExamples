## Info about the scikit-learn Iris dataset
it is packaged in scikit-learn
info on the data is available from https://scikit-learn.org/stable/datasets/index.html#iris-dataset
There are 3 classes and 50 instances of each class

Columns:
 
    sepal length in cm
    sepal width in cm
    petal length in cm
    petal width in cm

classes:
    Iris-Setosa
    Iris-Versicolour
    Iris-Virginica

Dataset:
Dictionary-like object, the interesting attributes are: 
- ‘data’, the data to learn
- ‘target’, the classification labels
- ‘target_names’ the meaning of the labels
- ‘feature_names’, the meaning of the features
- ‘DESCR’, the full description of the dataset
- ‘filename’, the physical location of iris csv dataset (added in version 0.20).

```python
# Load scikit-learn data
iris = datasets.load_iris()
```

```python
# Check out data
print("Num rows and Featues \n\t",iris.data.shape)
print("\nTypes of Iris \n\t",iris.target_names) # classification names
print("\nIris Features \n\t",iris.feature_names)
print("\n Iris data\n",iris.data[:5]) # check out some features
#print("\n All info in data \n\t",iris)
```

```python
X = iris.data[:, :2]  # we only take the first two features.
```

```python
# routine to plot 
def doplot(data, target, featurenames):
    X = data[:, :2]  # we only take the first two features.
    y = target
    f = featurenames

    plt.figure(2, figsize=(8, 6))
    plt.clf()
    
    # column 0 and 1
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
                edgecolor='k')
    z= str(y[0])
    plt.xlabel(f[0])
    plt.ylabel(f[1])

    #plt.xlim(x_min, x_max)
    #plt.ylim(y_min, y_max)
    plt.xticks(())
    plt.yticks(())
    plt.show()
```
